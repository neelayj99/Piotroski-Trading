```python
#2020
import yfinance as yf
import pandas as pd
li=[]
stock=input("Enter name of the Ticker to find the score ")
li.append(stock)
for i in li:
    company=i
    stock=yf.Ticker(company)
    year=2020
    score=0
    try:
        net_income1=stock.financials.loc['Net Income']['2020']
        net_income2=stock.financials.loc['Net Income']['2019']

        asset1=stock.balance_sheet.loc['Total Assets']['2020']
        asset2=stock.balance_sheet.loc['Total Assets']['2019']
        asset3=stock.balance_sheet.loc['Total Assets']['2018']

        cfo=stock.cashflow.loc['Total Cash From Operating Activities']['2020']

        longterm_debt1=stock.balance_sheet.loc['Long Term Debt']['2020']
        longterm_debt2=stock.balance_sheet.loc['Long Term Debt']['2019']

        current_assets1=stock.balance_sheet.loc['Total Current Assets']['2020']
        current_assets2=stock.balance_sheet.loc['Total Current Assets']['2019']
        current_assets3=stock.balance_sheet.loc['Total Current Assets']['2018']

        current_liabilities1=stock.balance_sheet.loc['Total Current Liabilities']['2020']
        current_liabilities2=stock.balance_sheet.loc['Total Current Liabilities']['2019']

        gross_profit1=stock.financials.loc['Gross Profit']['2020']
        gross_profit2=stock.financials.loc['Gross Profit']['2019']

        revenue1=stock.financials.loc['Total Revenue']['2020']
        revenue2=stock.financials.loc['Total Revenue']['2019']

        avg_current_assets1=((float(current_assets1)+float(current_assets2))/2)
        avg_current_assets2=((float(current_assets3)+float(current_assets2))/2)

        avg_assets1=((float(asset1)+float(asset2))/2)
        avg_assets2=((float(asset3)+float(asset2))/2)

        cash_flow_op=float(cfo)/float(avg_assets1)

        roa1=float(net_income1)/float(avg_assets1)
        roa2=float(net_income2)/float(avg_assets2)
        delta_roa=float(roa1-abs(roa2))

        accruals=float(roa1-abs(cash_flow_op))

        leverage1=float(longterm_debt1)/float(avg_assets1)
        leverage2=float(longterm_debt2)/float(avg_assets2)
        delta_leverage=float(leverage1-abs(leverage2))

        current_ratio1=((float(current_assets1)+float(current_assets2))/2)/float(current_liabilities1)
        current_ratio2=((float(current_assets3)+float(current_assets2))/2)/float(current_liabilities2)
        delta_liquidity=current_ratio1-abs(current_ratio2)

        gmo1=float(gross_profit1)/float(revenue1)
        gmo2=float(gross_profit2)/float(revenue2)
        delta_gmo=gmo1-abs(gmo2)

        atr1=float(revenue1)/float(avg_assets1)
        atr2=float(revenue2)/float(avg_assets2)
        atr=float(atr1-abs(atr2))
    except:
        print("Couldn't fetch Data")

    op=''
    try:
        print(stock.splits['2020'])
        op='Y'
    except:
        op='N'


    
    print("\n*************************************************Piotroski F-Score*************************************************\n")
    try:
        print(stock.info['shortName'])
        print(f"The ROA of the company for the year {year} is {round(roa1,3)}")
        print(f"The CFO of the company for the year {year} is {round(cash_flow_op,3)}" )
        print(f"The Delta ROA of the company for the year {year} is {delta_roa}" )
        print(f"The Accruals of the company for the year {year} is {accruals}" )
        print(f"The Delta Liquidity of the company for the year {year} is {delta_liquidity}" )
        print(f"The Delta Leverage of the company for the year {year} is {delta_leverage}" )
        print(f"The Delta Gross Margin of the company for the year {year} is {delta_gmo}" )
        print(f"The Delta Asset Turnover Ratio of the company for the year {year} is {atr}\n" )

        print(f"\nThe Current Ratio of the company for the year {year} is {current_ratio1}" )
        print(f"The Leverage of the company for the year {year} is {leverage1}" )
        print(f"The Gross Margin of the company for the year {year} is {gmo1}" )
        print(f"The Asset Turnover Ratio of the company for the year {year} is {atr1}" )
    except:
        print("Cannot calculate the score while displaying ")



    if(round(roa1,2)>0):
        score=score+1
    if(round(cash_flow_op,2)>0):
        score=score+1
    if(delta_roa>0):
        score=score+1
    if(accruals<0):
        score=score+1
    if(delta_leverage<0):
        score=score+1
    if(delta_liquidity>0):
        score=score+1
    if(delta_gmo>0):
        score=score+1
    if(atr>0):
        score=score+1
    if(op=="Y"):
        score=score+1


    print("The Piotroski F-Score is ", score)
    if(score>=8):
        print("It is expected of you to long this stock ")
    if(score<=2):
        print("It is expected of you to short this stock ")
```
